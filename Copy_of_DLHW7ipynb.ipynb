```
import keras

from keras import layers
from numpy import array
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
import numpy as np
import pandas as pd

import tensorflow as tf
```

```
def rec_fib(n):
    '''inefficient recursive function as defined, returns Fibonacci number'''
    if n > 1:
        return rec_fib(n-1) + rec_fib(n-2)
    return n
```

```

sequence_fib = [rec_fib(i) for i in range(1,13)]
x = sequence_fib[:-1]
y = sequence_fib[1:]
```

```
# x = np.array([sequence_fib[:-1]])
# y = np.array([sequence_fib[1:]])
# print(x.shape)
# print(y.shape)

x = array(x).reshape(11, 1, 1)
y = array(y).reshape(11,1,1)
```

```

def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```
x_normalized = NormalizeData(x)
```

```
y_normalized = NormalizeData(y)
```

```
print(x_normalized.shape)
y_normalized.shape
```

```
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=((1, 1))))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
print(model.summary())
```

```
model.fit(x_normalized, y_normalized, epochs=200)
```

```
test_input = array([144])
test_input = test_input.reshape((1, 1, 1))
test_output = model.predict(test_input)
print(test_output)
```
